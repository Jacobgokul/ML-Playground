##What Are Evaluation Metrics?

Evaluation metrics are used to measure the performance of a machine learning model. These metrics help us understand how well the model predicts outcomes and guide us in improving it.

## Why Are They Important?

- Helps compare different models.

- Identifies errors in predictions.

- Assists in model optimization.

#Types of Evaluation Metrics
##1. Regression Metrics (For Continuous Values)
- Mean Absolute Error (MAE) → Measures average absolute errors.

- Mean Squared Error (MSE) → Measures squared errors, penalizing larger mistakes.

- Root Mean Squared Error (RMSE) → Similar to MSE but in the same unit as the target.
- R² Score (Coefficient of Determination) → Measures how well the model explains variance.

##2. Classification Metrics (For Categorical Outcomes)
- Confusion Matrix → Shows true/false positives & negatives.
- Accuracy → Measures how many predictions are correct.

- Precision → Focuses on how many positive predictions were actually correct.

- Recall (Sensitivity) → Measures how well we identified actual positives.

- F1-Score → Balances precision and recall.

- ROC-AUC Score → Measures how well the model differentiates between classes.

# How to find variance
Consider 2 column in dataset

## Step 1: Compute Mean of Each Column
 - Mean of 1st column
 - mean_of_1stcol = sum(column value) / len(col value)

 - Mean of 2nd column
 - mean_of_2stcol = sum(column value) / len(col value)

## Step 2: Compute Variance of Each Column
 - variance of 1st column
 - formula: (1st_col_1st_val - mean_of_1stcol)^2 ... + (1st_col_n_val - mean_of_1stcol)^2 / len(1stcol)

  - variance of 2st column
 - formula: (2st_col_1st_val - mean_of_2stcol)^2 ... + (2st_col_n_val - mean_of_2stcol)^2 / len(2stcol)

# Classification Metrics

# **Accuracy**
- Measures the percentage of correctly classified samples.
- Formula:
     Accuracy = correct prediction / total prediction

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")
```

# **Confusion Matrix**
- Shows true positives (TP), false positives (FP), false negatives (FN), and true negatives (TN).
- Used to visualize classification performance.

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns

cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```

# **Precision**
- Measures how many positive predictions were actually correct.
- Important for imbalanced datasets (e.g., detecting fraud where positives are rare).
- Formula
  Precision = TP / TP + FP

```python
from sklearn.metrics import precision_score
precision = precision_score(y_test, y_pred)
print(f"Precision: {precision}")
```

# **Recall**
- Measures how many actual positive samples were correctly classified.
- Important when false negatives are costly (e.g., medical diagnoses).
- Formula:
     Recall = TP / TP + FN

```python
from sklearn.metrics import recall_score
recall = recall_score(y_test, y_pred)
print(f"Recall: {recall}")
```

# **F1-Score**
- Harmonic mean of Precision & Recall.
- Balances false positives and false negatives.
- Formula: f1 = 2 * (precision * recall / precision + recall)

```python
from sklearn.metrics import f1_score
f1 = f1_score(y_test, y_pred)
print(f"F1 Score: {f1}")
```

# **ROC-AUC Score (for Probabilistic Models)**
- Measures the model’s ability to distinguish between classes.
- AUC = 1 (perfect model), AUC = 0.5 (random guessing).





```python
from sklearn.metrics import roc_auc_score
auc = roc_auc_score(y_test, y_pred_prob)
print(f"ROC-AUC Score: {auc}")
```
