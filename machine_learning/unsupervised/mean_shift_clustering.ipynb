### What is Mean Shift?

- A density-based clustering algorithm.

- Finds peaks (high-density regions) in the data.

- Each peak becomes a cluster center.

- Unlike K-Means:

    - No need to specify number of clusters.

    - It automatically finds how many clusters exist.

#### Key Concept

- Every point looks around itself in a window (called bandwidth).

- It calculates the mean (average) of points inside that window.

- Then it shifts towards that mean.

- After many shifts, points gather around density peaks → clusters form.

#### How it Works
- Place a window (circle) around a data point.

- Compute the mean of all points inside the window.

- Move the window’s center to that mean (shift).

- Repeat until the center does not move anymore.

- Points that end up at the same peak = same cluster.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import MeanShift
from sklearn.datasets import make_blobs

# Generate sample data
X, _ = make_blobs(n_samples=300, centers=3, cluster_std=0.8, random_state=42)
```

```python
X
```

```python
# Apply Mean Shift
ms = MeanShift(bandwidth=2)   # bandwidth = window size
labels = ms.fit_predict(X)
centers = ms.cluster_centers_
centers
```

```python
# Plot results
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap="plasma", s=30)
plt.scatter(centers[:, 0], centers[:, 1], c="black", marker="x", s=200)
plt.title("Mean Shift Clustering")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

Output:

- Same colors = same cluster.

- Black “X” = cluster centers (density peaks).
