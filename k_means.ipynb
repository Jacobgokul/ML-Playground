# What is Clustering?
- It groups similar data points together.
- Each group is called a cluster.
- Used in customer segmentation, fraud detection, document categorization, etc.

# What is K-means?
K-Means is an unsupervised clustering algorithm that groups data into K distinct clusters based on similarity. It minimizes the variance within each cluster while maximizing the variance between clusters.


# How K-Means Works
##Step 1: Select the Number of Clusters (K)

The user must choose K, the number of clusters.

If K is not known, the Elbow Method or Silhouette Score can help determine it.

##Step 2: Initialize K Centroids

Randomly select K points (centroids) from the dataset.

These centroids act as the starting points for the clusters.

##Step 3: Assign Each Data Point to the Nearest Centroid

Calculate the Euclidean distance between K and clusters.

Assign each point to the closest centroid, forming K clusters.

image.png

##Step 4: Compute New Centroids

After assignment, compute the new centroid of each cluster:

Take the mean of all points in a cluster.

Update the centroid position.

##Step 5: Repeat Until Convergence

Repeat Steps 3 & 4 until:

Centroids do not change (or change very little).

A maximum number of iterations is reached.

##Step 6: Final Clusters

The algorithm outputs K clusters where each data point belongs to a specific group.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
```

```python
# Generate synthetic data with 3 clusters
X, y = make_blobs(n_samples=300, centers=3, cluster_std=1.05, random_state=42)
```

```python
# Scatter plot of the dataset
plt.scatter(X[:, 0], X[:, 1], s=50, alpha=0.6)
plt.title("Generated Dataset for K-Means Clustering")
plt.show()
```

```python
y
```

```python
# Apply K-Means clustering with K=3
kmeans = KMeans(n_clusters=3, init='k-means++', max_iter=300, random_state=42)
y_kmeans = kmeans.fit_predict(X)
```

```python
y_kmeans
```

```python
# Get cluster centers | Retrieves the coordinates of the final cluster centers.
centroids = kmeans.cluster_centers_
centroids
```

```python
# Visualizing the clustered data
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s=50, c='red', label='Cluster 1')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s=50, c='blue', label='Cluster 2')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s=50, c='green', label='Cluster 3')
```

```python
# Plot cluster centroids
plt.scatter(centroids[:, 0], centroids[:, 1], s=200, c='yellow', marker='X', edgecolors='black', label='Centroids')
```

```python
# New data points to classify
new_points = np.array([[2, 3], [-4, 7], [6, -2]])
# Predict the cluster for new data points
predicted_clusters = kmeans.predict(new_points)

# Print results
for i, point in enumerate(new_points):
    print(f"Point {point} belongs to Cluster {predicted_clusters[i]}")
```

```python
# Plot the existing clusters
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s=50, c='red', label='Cluster 1')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s=50, c='blue', label='Cluster 2')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s=50, c='green', label='Cluster 3')

# Plot centroids
plt.scatter(centroids[:, 0], centroids[:, 1], s=200, c='yellow', marker='X', edgecolors='black', label='Centroids')

# Plot new points with their predicted clusters
plt.scatter(new_points[:, 0], new_points[:, 1], s=150, c='purple', marker='D', label='New Points')

plt.legend()
plt.title("K-Means Clustering with Predictions")
plt.show()
```
